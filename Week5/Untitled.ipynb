```python
from PIL import Image, ImageDraw, ImageFont

# Create a new image with white background
image = Image.new("RGB", (400, 350), "white")

# Open and resize img1 and img2
img1 = Image.open("img1.jpg")
img2 = Image.open("img2.jpg")
img1 = img1.resize((200, 200))
img2 = img2.resize((200, 200))

# Paste img1 and img2 side by side
image.paste(img1, (0, 0))
image.paste(img2, (img1.width, 0))

# Create a new image for the text background
text_bg = Image.new("RGB", (165, 95), (187, 183, 166))

# Create a new draw object for the text background image
draw_text_bg = ImageDraw.Draw(text_bg)

# Set the font for the text
font = ImageFont.truetype("arial.ttf", 16)
font1 = ImageFont.truetype("arialbi.ttf",26)

# Draw the text onto the text background image
text = "Online Yoga\nwith Finance\nClass"
draw_text_bg.multiline_text((5, 5), text, fill="white", font=font1)

# Paste the text background image onto the main image
image.paste(text_bg, (135, img1.height-50))

# Open and resize img3
img3 = Image.open("img3.jpg")
img3 = img3.resize((60, 60))

# Paste img3 below the text background image
image.paste(img3, (305, img1.height + 15))

# Create a new draw object for the text on the main image
draw = ImageDraw.Draw(image)

# Draw the "How to Join Yoga Class:" text in the first partition
bold_font = ImageFont.truetype("arialbd.ttf", 16)
draw.text((10, img1.height + 50), "How to Join Yoga Class:", fill="black", font=bold_font)
draw.text((10, img1.height + 70), "Step-1. Join training program", fill="black", font=font)
draw.text((10, img1.height + 90), "Step-2. Pay Rs.1, Plant a Tree", fill="black", font=font)

# Draw the "Monthly Fee : Rs 1" text in the first partition, to the right of other text
draw.text((260, img1.height + 80), "Monthly Fee : Rs 1", fill="black", font=bold_font)

# Create a new image for the colored background of the "Join Now" text
join_now_bg = Image.new("RGB", (80, 20), (187, 183, 166))

# Paste the join now background image onto the main image
image.paste(join_now_bg, (295, img1.height + 100))

# Draw the "Join Now" text on the join now background
draw.text((300, img1.height+100), "Join Now", fill="white", font=bold_font)

# Save the final image
image.save("output.jpg")
```
