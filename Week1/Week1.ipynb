### 4.Extracting phone number from JSON file and save into ms excel

```python
import json
import re
from openpyxl import load_workbook

# Open the JSON file and load the data
with open('chat1.json') as f:
    data = json.load(f)
    
with open('chat2.json') as f2:
    data2 = json.load(f2)
    
with open('chat3.json') as f3:
    data3 = json.load(f3)

# Load the Excel file and select the active sheet
workbook = load_workbook(filename='my_excel_file.xlsx')
sheet = workbook.active

# Extract all the mobile numbers from the messages
mobile_numbers = []
for message in data['messages']:
    if 'msg' in message and message['msg']:
        match = re.search(r'Phone\s*:\s*(\d+)', message['msg'])
        if match:
            mobile_numbers.append(match.group(1))
            
for message in data2['messages']:
    if 'msg' in message and message['msg']:
        match = re.search(r'Phone\s*:\s*(\d+)', message['msg'])
        if match:
            mobile_numbers.append(match.group(1))
            
for message in data3['messages']:
    if 'msg' in message and message['msg']:
        match = re.search(r'Phone\s*:\s*(\d+)', message['msg'])
        if match:
            mobile_numbers.append(match.group(1))
            
# Write the mobile numbers to the Excel file
for i, number in enumerate(mobile_numbers):
    cell = sheet.cell(row=i+1, column=1)
    cell.value = number

# Save the changes to the Excel file
workbook.save('my_excel_file.xlsx')

#printing mobile numbers
print(mobile_numbers)
```
